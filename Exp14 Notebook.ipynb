```python
import pandas as pd
import numpy as np

data = {
    'Product':['Laptop','Mobile', 'Tablet', 'Headphones', 'Smartwatch', 'Camera'],
    'Price': [55000,20000,30000,2000,8000,45000],
    'Units_Sold': [25,60,35,80,50,15]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Data Binning
bins1 = [0,10000,30000,60000]
labels1 = ['Low', 'Medium', 'High']

df['Price_Category'] = pd.cut(df['Price'], bins=bins1, labels=labels1)
df
```

```python
#Data binnig sold
bins2 = [0,30,60,100]
labels2 = ['Low sales', 'Medium Sales', 'High sales']

df['Sales_Category'] = pd.cut(df['Units_Sold'], bins=bins2, labels=labels2)
df
```

```python
# data formating
#convert data type

print(df.dtypes)
df['Units_Sold'] = df['Units_Sold'].astype(float)
print(df.dtypes)
df
```

```python
#convert text to uppercase
df['Product'] = df['Product'].str.upper()
df
```

```python
# Rounding values
df['Price'] = df['Price'].round(2)

df
```

```python
#additional operation
#Sort data

sorted_df = df.sort_values(by='Price')
print(sorted_df)
```

```python

df.sort_values(by='Units_Sold', inplace = True)
print(df)
```

```python
df.sort_values(by=['Units_Sold'], inplace = True, ascending = False)
print(df)
```

```python
#Display Unique Values
print(df['Price_Category'].unique())
```

```python
import pandas as pd
import numpy as np
data = {
    'Order_ID': ['01','02','03','04','05','06','07','08'],
    'Order_Value': [250,450,120,600,300,150,500,220],
    'Delivery_Time':[30,45,25,60,35,20,50,28],
    'Distance_Km':[5,8,3,10,6,2,9,4]

}
df = pd.DataFrame(data)
df
```

```python
#Binning oder values

bins=[0,200,400,700]
labels = ['Low Order', 'Medium Order', 'High Order']

df['Order_Category'] = pd.cut(df['Order_Value'], bins=bins, labels=labels)
df
```

```python
bins2 = [0,30,45,70]
labels2 = ['Fast Delivery','Normal Delivery','Slow Delivery']
df['Delivery_Category'] = pd.cut(df['Delivery_Time'],bins=bins2,labels=labels2)
df
```

```python
bins3 = [0,4,8,12]
labels3 = ['far','moderately far','close']
df['Distance_Category'] = pd.cut(df['Distance_Km'],bins=bins3,labels=labels3)
df
```

```python
print(df['Order_Category'].value_counts())
df
```
